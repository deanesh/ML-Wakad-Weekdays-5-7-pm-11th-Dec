Prompt: 
Generate data for Microsoft, Tesla, Apple for Fiscal Years 2022, 2023 and 2024 in csv format                  [Company, Fiscal Year,Total Revenue,Net Income,Total Assets,Total Liabilities,Cash Flow from Ops]

```python
import pandas as pd
```

```python
# Reading 10-K Filings for Years 2022, 2023 and 2024
df = pd.read_csv('10-K-Filings-Last-3-Fiscal-Years.csv')
```

```python
df['Revenue Growth (%)'] = round(df.groupby(['Company'])['Total Revenue'].pct_change() * 100, 2)
df['Net Income Growth (%)'] = round(df.groupby(['Company'])['Net Income'].pct_change() * 100, 2)
print('Revenue Growth (%):', df['Revenue Growth (%)'].tolist())
print('Net Income Growth (%):', df['Net Income Growth (%)'].tolist())
```

```python
def simple_chatbot(user_query):
    user_query = user_query.lower()

    if "total revenue in 2024 for apple" in user_query:
        revenue = df[(df["Company"] == "Apple Inc.") & (df["Fiscal Year"] == 2024)]["Total Revenue"].values[0]
        return f"Apple's total revenue in 2024 was {revenue} million USD."
    elif "revenue changed for microsoft from 2023 to 2024" in user_query:
        rev_2023 = 211915
        rev_2024 = 245122
        change = rev_2024 - rev_2023
        return f"Microsoft's revenue increased by {change} million USD from 2023 to 2024."
    elif "tesla's net income in 2023" in user_query:
        return "Tesla's net income in 2023 was 14997 million USD."
    elif "tesla's net income in 2022" in user_query:
        return "Tesla's net income in 2022 was 12587 million USD."
    elif "apples's net income in 2022" in user_query:
        return "Apple's net income in 2022 was 99803 million USD."
    elif "highest revenue in 2024" in user_query:
        highest = df.loc[df["Total Revenue"].idxmax()]
        return f"{highest['Company']} had the highest revenue in 2024 with {highest['Total Revenue']} million USD."
    elif "operating cash flow of apple in 2024" in user_query:
        cash_flow = df[(df["Company"] == "Apple Inc.") & (df["Fiscal Year"] == 2024)]["Cash Flow from Ops"].values[0]
        return f"Apple's operating cash flow in 2024 was {cash_flow} million USD."
    else:
        return "Sorry, I can only answer predefined financial questions."

```

```python
print("Financial Chatbot (type 'exit' to quit)")

while True:
    query = input("You: ")
    if query.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    response = simple_chatbot(query)
    print("Chatbot:", response)
```
