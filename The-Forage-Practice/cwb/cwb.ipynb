```python
import pandas as pd
df = pd.read_excel('supermarket_transactions.xlsx')

# Q1: Across locations, how many apples were purchased in cash?
apples_purchased_in_cash = df.loc[
    (df['product_name'] == 'apple') &
    (df['payment_method'] == 'cash')
]

total_apples_purchased = apples_purchased_in_cash['quantity'].sum()
print(f"Total apples purchased using cash: {total_apples_purchased}")

# Q2: How much total cash was spent on these apples?
total_cash_spent_on_apples = apples_purchased_in_cash['total_amount'].sum()
print(f"Total cash spent on apples: ${total_cash_spent_on_apples:.2f}")

# Q3: Across all payment methods, how much money was spent at the Bakershire store by non-member customers?
bakershire_non_members = df.loc[
    (df['store'] == 'Bakershire') &
    (df['customer_type'] == 'non-member')
]

total_spent_bakershire_non_members = bakershire_non_members['total_amount'].sum()
print(f"Total spent by non-members at Bakershire: ${total_spent_bakershire_non_members:.2f}")

```
