```python

# ----------------------------
# 2. Planning Agent (Mock)
# ----------------------------
print("--- Planning Agent Demo ---")
class MockPlanner:
    def plan(self, goal):
        return [f"Step-{i}: Towards {goal}_{i}\n" for i in range(1, 4)]

planner = MockPlanner()
plan = planner.plan("picking up object")
print("Planning: Start plan")
print("".join(plan))
print("Planning: Plan complete")
print("--- End of Planning Agent ---\n")
```
