| Aspect                 | Description |
|------------------------|-------------|
| **Definition**         | AI systems that can **act autonomously** to achieve goals, make decisions, and adapt strategies without constant human intervention. |
| **Core Feature**       | **Agency** – ability to perceive, reason, plan, and take actions in dynamic environments. |
| **Examples of Tasks**  | Task automation, autonomous agents in games, robotic control, digital assistants, decision-making systems. |
| **Types of Models**    | 1. **Reinforcement Learning (RL) Agents** – Learn via trial and error (e.g., AlphaZero).<br>2. **Planning/Decision Agents** – Use search and optimization to plan actions (e.g., STRIPS planners).<br>3. **Multi-Agent Systems** – Interacting agents that cooperate or compete (e.g., autonomous vehicle fleets).<br>4. **Language-based Agents** – LLMs with action tools, capable of reasoning and performing tasks (e.g., AutoGPT, LangChain agents). |
| **Key Components**     | - **Perception:** Sensing environment<br>- **Reasoning/Planning:** Deciding next action<br>- **Action Execution:** Performing tasks<br>- **Feedback:** Learning from outcomes |
| **Applications**       | Robotics, autonomous vehicles, virtual assistants, strategic games, simulation management, workflow automation |
| **Challenges**         | Safety, alignment with human goals, interpretability, adaptability, multi-agent coordination |
