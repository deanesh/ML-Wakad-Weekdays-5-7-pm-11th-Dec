```python
# ----------------------------
# 4. Language-based Agent (Mock Tool)
# ----------------------------
print("--- Language-based Agent Demo ---")
class MockLanguageAgent:
    def search(self, query):
        return f"Mock search result for '{query}'"
    def run(self, query):
        return self.search(query)

agent = MockLanguageAgent()
print("Language Agent: Start query")
result = agent.run("Find the population of France")
print("Language Agent Result:", result)
print("--- End of Language-based Agent ---\n")
```
