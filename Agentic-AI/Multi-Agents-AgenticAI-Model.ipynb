```python
# ----------------------------
# 3. Multi-Agent System (PettingZoo)
# ----------------------------

from warnings import filterwarnings
filterwarnings('ignore')
from pettingzoo.mpe import simple_spread_v3
print("--- Multi-Agent System Demo ---")
env = simple_spread_v3.env(N=3)
env.reset()

print("Multi-Agent: Start timestep")
for i, agent in enumerate(env.agent_iter(max_iter=4)):
    obs, reward, terminated, truncated, info = env.last()
    done = terminated or truncated
    action = env.action_space(agent).sample()
    env.step(action)
    print(f"Timestep {i+1}: Agent {agent} action={action}, reward={reward:.2f}")
env.close()
print("Multi-Agent: End timestep")
print("--- End of Multi-Agent System ---\n")

```
