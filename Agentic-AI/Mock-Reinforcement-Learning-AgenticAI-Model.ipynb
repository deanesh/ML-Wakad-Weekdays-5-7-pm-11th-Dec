```python
# ============================
# Agentic AI Hybrid Demo (Compact Output)
# ============================

from stable_baselines3 import PPO
from warnings import filterwarnings
filterwarnings('ignore')
import gym
import numpy as np

# Patch for NumPy 2.x
if not hasattr(np, "bool8"):
    np.bool8 = np.bool_

print("=== Agentic AI Demo (Compact Output) ===\n")
# ----------------------------
# 1. RL Agent (Gym + SB3)
# ----------------------------
print("--- RL Agent Demo ---")
env = gym.make("CartPole-v1")
model = PPO("MlpPolicy", env, verbose=0)
model.learn(total_timesteps=500)  # very short training for compact output

obs, info = env.reset()
print("RL: Start state:", obs)
for i in range(5):  # only 5 steps printed
    action, _ = model.predict(obs)
    obs, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated
    print(f"RL Step {i+1}: action={action}, reward={reward:.2f}")
    if done:
        obs, info = env.reset()
print("RL: End state:", obs)
env.close()
print("--- End of RL Agent ---\n")
```
