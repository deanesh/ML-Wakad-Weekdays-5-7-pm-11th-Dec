##### <b>`Forward Feature Selection`<b>
###### <b>Problem Stmt: Estimate Weight (Col) based on other Cols in data (CSV File)<b>

```python
# Ignore Warnings
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
#Step-1: Data Ingestion
import pandas as pd
df = pd.read_csv('Cars93.csv', keep_default_na=False, na_values=["", "NA"])
df.head(1)
```

```python
#Step-2: Data Sanity - Duplicate removals
```

```python
# Duplicate Data
duplicate_count = df.duplicated().sum()

if duplicate_count > 0:
    print(f'Duplicates Found: {duplicate_count}')
    print('Removing Duplicates...')
    df = df.drop_duplicates(keep = "first").reset_index(drop = True)
    print('Removed Duplicates...')
    df.shape
else:
    print('No Duplicates Found')
```

```python
#Step-3: Separate X and Y
X = df.drop(columns = ["id", "Weight"])
Y = df["Weight"]
X.head(1)
```

```python
#Step-4: Remove High Unique Cat Cols
card = df.select_dtypes(include = "object").nunique() / len(df)
X = X.drop(columns = card[card >= 0.9].index)
```

```python
#Step-5: Train Test Split
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Step-6: Apply Preprocessing on X
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy = "median"),
    StandardScaler()
)

cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1),
    StandardScaler()
)

num_cols = X.select_dtypes(include = "number").columns
cat_cols = X.select_dtypes(include = "object").columns

pre = ColumnTransformer(
    [
        ("num", num_pipe, num_cols),
        ("cat", cat_pipe, cat_cols)
    ]
).set_output(transform = "pandas")

pre.fit(xtrain)

xtrain_pre = pre.transform(xtrain)
xtest_pre = pre.transform(xtest)
```

```python
#Step-7: Forward Feature Selection
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression

base_model = LinearRegression()
seq_feature_selector = SequentialFeatureSelector(
    base_model, n_features_to_select="auto", scoring="r2", direction="forward"
)
seq_feature_selector.fit(xtrain_pre, ytrain)
sel_cols = seq_feature_selector.get_feature_names_out()
print(f'[Sequential Feature Selector] Selected Features Count: {len(sel_cols)}')

xtrain_pre_sel = seq_feature_selector.transform(xtrain_pre)
xtest_pre_sel = seq_feature_selector.transform(xtest_pre)
```

```python
#Step-8: Build Model
from sklearn.linear_model import LinearRegression

model_1 = LinearRegression()
model_1.fit(xtrain_pre_sel, ytrain)

print(f'[Forward Feature Selection Model] Train Score : {model_1.score(xtrain_pre_sel, ytrain)}')
print(f'[Forward Feature Selection Model] Test Score :{model_1.score(xtest_pre_sel, ytest)}')
```

```python
#Step-9: Evaluate Model
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)

def evaluate_metrics(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"[RMSE : {rmse:.2f}| MAE : {mae:.2f}| MAPE : {mape:.2%}| R2 : {r2:.2%}]")

print("[Forward Feature Selection Model] Train Results :")
evaluate_metrics(model_1, xtrain_pre_sel, ytrain)
print("\n[Forward Feature Selection Model] Test Results :")
evaluate_metrics(model_1, xtest_pre_sel, ytest)
print(f'\n[Forward Feature Selection Model] Predicted Weights : {model_1.predict(xtest_pre_sel)[0:2]}')
```

```python
#Step-10: Model Inference (Out of Sample Prediction)
xnew = pd.read_csv("sample.csv", na_values = ["", "NA"], keep_default_na = False)
xnew = xnew.drop(columns = card[card >= 0.9].index)
xnew_pre = pre.transform(xnew)
xnew_pre_sel = seq_feature_selector.transform(xnew_pre)
preds = model_1.predict(xnew_pre_sel)
print(f'\n[Forward Feature Selection Model]--Out of Sample: Predicted Weights(First 2) : {preds[0:2]}')
```
