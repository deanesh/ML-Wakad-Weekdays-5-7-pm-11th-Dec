**Question 5 — Thermodynamics**

One mole of an ideal monatomic gas undergoes a cyclic process \(A \to B \to C \to A\) as shown in the \(V-T\) diagram. Process \(A \to B\) is isochoric, \(B \to C\) is linear in \(V-T\), and \(C \to A\) is isothermal. If the net work done by the gas in the cycle is \(nRT_0\), where \(T_0\) is the temperature at \(A\), then \(n\) is:

[Given data: \(T_A = T_0\), \(T_B = 2T_0\), \(T_C = 4T_0\), \(V_A = V_0\), \(V_B = V_0\), \(V_C = 2V_0\)]

(A) \(\ln 2 - 1.5\)  
(B) \(2 \ln 2 - 1.5\)  
(C) \(\ln 2 - 0.5\)  
(D) \(2 \ln 2 - 0.5\)

**Justification of Difficulty**

1. **Q1 (Modern Physics):** Requires using Rydberg formula for hydrogen-like atoms for multiple transitions, solving simultaneous equations for \(Z\) and energy levels. The two-step emission with given wavelengths tests systematic understanding of energy levels.

2. **Q2 (Electrodynamics):** Infinite LC ladder network impedance calculation requires knowledge of iterative impedance formula for reactive networks. Solving \(Z = Z_C + (Z_L \parallel Z)\) leads to a quadratic in \(Z\). Complex number manipulation at given \(\omega\) is needed.

3. **Q3 (Mechanics):** Combines rolling without slipping on incline, translational+rotational energy, and vertical circular motion condition for rolling object. The transition to frictionless track changes constraints—must conserve angular momentum at the transition? Actually, since track is frictionless, cylinder will slip but its rotational energy remains constant during circular motion. Must compute velocity at top for just completing circle with rotational kinetic energy present.

4. **Q4 (Optics):** Involves Newton's rings formula, path difference change when lens is raised, and differentiation to relate change in diameter to change in air gap. Requires careful handling of phase change at reflections and bright ring condition.

5. **Q5 (Thermodynamics):** Mixed process in V-T coordinates, work calculation for each segment including linear process B→C via integration. Requires combining first law with ideal gas equation, and computing area in cycle correctly.
