(b) 1,4-Cyclohexadiene

(a) 10 min  
(b) 20 min  
(c) 40 min  
(d) 80 min

**Justification:** First order: t depends on log(C₀/C). Ratio 4 each time → t = 40 min.

<summary>Q7. Conductivity problem: KCl, AgNO₃. Solubility product of AgCl?</summary>

(a) 1.0 × 10⁻¹⁰  
(b) 2.5 × 10⁻¹⁰  
(c) 1.0 × 10⁻⁹

</details>

<details>
<summary>Q8. Concentration cell: Pt | H₂(1 atm) | H⁺ (pH=3) || H⁺ (pH=5) | H₂(1 atm) | Pt. EMF?</summary>

(a) 0.059 V  
(b) 0.118 V  
(c) -0.059 V  
(d) -0.118 V

**Justification:** E = 0.059 log([H⁺]₂/[H⁺]₁) = 0.059 log(10⁻⁵/10⁻³) = -0.118 V. Adjust sign according to anode/cathode.

</details>

<summary>Q9. 2A → products, half-life inversely proportional to [A]₀. Time to fall to 1/16th of [A]₀?</summary>

(a) 4

(c) 15  
(d) 16

**Justification:** Second order: t = 1/[A]₀ units → t = 15 t₁/₂.

</details>

<summary>Q10. E°: Fe³⁺/Fe²⁺=0.77V, Fe²⁺/Fe=-0.44V, Fe³⁺/Fe=-0.04V. ΔG° for 2Fe³⁺ + Fe → 3Fe²⁺?</summary>

(a) -21.9 kJ/mol  
(b) -43.8 kJ/mol

(d) -87.6 kJ/mol

</details>
