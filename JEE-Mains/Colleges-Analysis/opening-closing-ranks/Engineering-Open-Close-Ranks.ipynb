```python
# JEE-Mains 2025 ALL BRANCHES (GFTI/IIIT/NIT) Opening- Closing Ranks List (General Category)
# ------------------------------------------------------------------------------------------

import pandas as pd

print('JEE-Mains 2025 ALL BRANCHES Opening- Closing Ranks List (General Category)....')
print('=========================================================================')

df = pd.read_csv('gfti-colleges-open-close-ranks-2025.csv')
gfti_ranks_for_general_category = (
    df.query('`Seat Type` == "OPEN" and `Gender` == "Gender-Neutral" and `Quota` in ["AI", "OS"] ')
)
gfti_open_rank = int(gfti_ranks_for_general_category['Opening Rank'].min())
gfti_close_rank = int(gfti_ranks_for_general_category['Closing Rank'].max())

print("GFTI Colleges [ Min Opening Rank:", gfti_open_rank, '| Max Closing Rank:', gfti_close_rank,"]")

df = pd.read_csv('iiit-colleges-open-close-ranks-2025.csv')
iiit_ranks_for_general_category = (
    df.query('`Seat Type` == "OPEN" and `Gender` == "Gender-Neutral" and `Quota` in ["AI", "OS"] ')
)
iiit_open_rank = int(iiit_ranks_for_general_category['Opening Rank'].min())
iiit_close_rank = int(iiit_ranks_for_general_category['Closing Rank'].max())

print("IIIT Colleges [ Min Opening Rank:", iiit_open_rank, '| Max Closing Rank:', iiit_close_rank,"]")

df = pd.read_csv('nit-colleges-open-close-ranks-2025.csv')
nit_ranks_for_general_category = (
    df.query('`Seat Type` == "OPEN" and `Gender` == "Gender-Neutral" and `Quota` in ["AI", "OS"] ')
)
nit_open_rank = int(nit_ranks_for_general_category['Opening Rank'].min())
nit_close_rank = int(nit_ranks_for_general_category['Closing Rank'].max())

print("NIT Colleges [ Min Opening Rank:", nit_open_rank, '| Max Closing Rank:', nit_close_rank,"]")
```

```python
# JEE-Mains 2025 Mains IIIT Colleges Opening/Closing Ranks for General Category by Branches
# -----------------------------------------------------------------------------------------

gfti_df = pd.read_csv('gfti-colleges-open-close-ranks-2025.csv')

def get_course_df(df, course_name):
    mask = (
        (df['Seat Type'] == "OPEN") &
        (df['Gender'] == "Gender-Neutral") &
        (df['Quota'].isin(["AI", "OS"])) &
        (df['Academic Program Short Name'].str.contains(course_name, case=False, na=False, regex=False))
    )
    return df[mask]


print(f'\nJEE-Mains 2025 Mains GFIT Colleges Opening/Closing Ranks for General Category by Branches')
print('=' * 91)

for course_name in gfti_df['Academic Program Short Name'].unique():
    gfti_course_df = get_course_df(gfti_df, course_name)

    min_open = gfti_course_df["Opening Rank"].min()
    max_close = gfti_course_df["Closing Rank"].max()

    if pd.isna(min_open) or pd.isna(max_close):
        print("-")
        print(f"{course_name}: [None | None]")
    else:
        print("---------------------------------------------------")
        print(
            f"{course_name}: "
            f"[ Opening Rank: {int(min_open)} | "
            f" Closing Rank: {int(max_close)} ]"
        )
```
