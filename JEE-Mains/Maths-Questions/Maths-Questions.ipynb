##### Conic Lesson Related

**Question 1 — Coordinate Geometry (Conics)**  

A parabola \( y^2 = 4ax \) is given. A circle is drawn with its center at the vertex of the parabola and passing through the focus. This circle intersects the parabola again at points \( P \) and \( Q \). If the normals to the parabola at \( P \) and \( Q \) intersect at point \( R \), and \( \triangle PQR \) is equilateral, then the area of \( \triangle PQR \) is:

(A) \( 12\sqrt{3} \, a^2 \)  
(B) \( 18\sqrt{3} \, a^2 \)  
(C) \( 24\sqrt{3} \, a^2 \)  
(D) \( 36\sqrt{3} \, a^2 \)

**Question 2 — Integral Calculus**  

Let \( f(x) = \int_0^x e^{t^2}(t^4 - 2t^2 + 3) \, dt \). The number of points of local maxima of \( f(x) \) in the interval \( (-\infty, \infty) \) is:

(A) 0  
(B) 1  
(C) 2  
(D) 3

**Question 3 — Algebra (Complex Numbers + Quadratics)**  

Let \( z_1, z_2, z_3 \) be three distinct complex numbers such that \( |z_1| = |z_2| = |z_3| = 1 \) and \( z_1 + z_2 + z_3 = 0 \). If \( z_1^4 + z_2^4 + z_3^4 = 0 \), then the value of \( |z_1 - z_2|^2 + |z_2 - z_3|^2 + |z_3 - z_1|^2 \) is:

(A) 3  
(B) 6  
(C) 9  
(D) 12

**Question 4 — Vector & 3D Geometry**  

The lines \( \frac{x-1}{2} = \frac{y-2}{3} = \frac{z-3}{4} \) and \( \frac{x-2}{3} = \frac{y-3}{4} = \frac{z-4}{5} \) are coplanar. A plane containing these lines also passes through the point \( (\alpha, \beta, \gamma) \) where \( \alpha + \beta + \gamma = 6 \). The equation of a sphere with center \( (1, 2, 3) \) and touching this plane is:

(A) \( (x-1)^2 + (y-2)^2 + (z-3)^2 = 1 \)  
(B) \( (x-1)^2 + (y-2)^2 + (z-3)^2 = 4 \)  
(C) \( (x-1)^2 + (y-2)^2 + (z-3)^2 = 9 \)  
(D) \( (x-1)^2 + (y-2)^2 + (z-3)^2 = 16 \)  

---

**Justification of Difficulty**

1. **Q1 (Conics):** Combines parabola, circle through focus and vertex, normals meeting, and equilateral triangle condition — needs parametric coordinates manipulation, properties of normals, and geometry of normals intersection, leading to solving for parameter \( a \) and area calculation.
   
2. **Q2 (Integral Calculus):** Involves differentiation under integral sign (Leibniz) to find maxima of an integral function with exponential and polynomial integrand; requires solving \( f'(x)=0 \) and analyzing sign changes of derivative, which reduces to solving \( e^{x^2}(x^4 - 2x^2 + 3) = 0 \) but careful: derivative \( f'(x) = e^{x^2}(x^4 - 2x^2 + 3) \) is always positive, hence no maxima — tricky to realize without computation.

3. **Q3 (Complex Numbers):** Conditions imply roots of unity properties; \( z_1, z_2, z_3 \) vertices of equilateral triangle on unit circle, additional condition \( z_1^4 + z_2^4 + z_3^4 = 0 \) must be consistent. Then sum of squares of side lengths needed; yields a constant independent of orientation — involves symmetry and algebra of cube roots of unity or \( 120^\circ \) separation.

4. **Q4 (3D Geometry):** Requires checking coplanarity, finding plane equation containing two skew lines, using given point condition to fix plane parameter, then distance from center to that plane is radius of sphere — multi-step with analytical geometry and sphere equation.
