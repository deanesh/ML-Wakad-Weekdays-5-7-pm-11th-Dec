## Q1. (Conics – Parabola + Directrix)

A parabola has focus at $(2,3)$ and directrix  
$$
x + y = 1.
$$

A circle with center $(a,b)$ and radius $r$ touches the directrix and also touches the parabola at its vertex.  
Then the value of  
$$
a^2 + b^2 + r^2
$$
is:

(A) 12  
(B) 18  
(C) 24  
(D) 30

## Q2. (Conics – Ellipse and Hyperbola, Common Foci)

The ellipse  
$$
\frac{x^2}{25} + \frac{y^2}{16} = 1
$$

and the hyperbola  
$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$$

have the same foci. If the eccentricity of the hyperbola is $3/2$, then the area of the quadrilateral formed by the asymptotes of the hyperbola and the lines  
$$
x = -5 \quad \text{and} \quad x = 5
$$
is:

(A) 30  
(B) 40  
(C) 50  
(D) 60

$$

$$
x^2 + y^2 = 4.
$$

(A) $x^2 + y^2 = 8$  
(B) $x^2 + y^2 = 16$  
(C) $x^2 + y^2 = 32$

## Q10. (Vectors – Zero Sum Identity)

$$
|\vec{a}| = 2,\quad |\vec{b}| = 3,\quad |\vec{c}| = 4
$$
and  
$$
\vec{a} + \vec{b} + \vec{c} = \vec{0}.
$$

The value of  
$$
\vec{a}\cdot\vec{b} + \vec{b}\cdot\vec{c} + \vec{c}\cdot\vec{a}
$$
is:

(A) $-29/2$  
(B) $-25/2$  
(C) $-21/2$  
(D) $-17/2$
